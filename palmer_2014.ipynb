# Palmer2014 - Effect of IL-1β-Blocking therapies in T2DM - Disease Condition

https://www.ebi.ac.uk/biomodels/BIOMD0000000620

Ref: https://ascpt.onlinelibrary.wiley.com/doi/pdfdirect/10.1038/psp.2014.16 - Effects of IL-1β–Blocking Therapies in Type 2 Diabetes Mellitus: A Quantitative Systems Pharmacology Modeling Approach to Explore Underlying Mechanisms

```python
import tellurium as te
```

```python
model = te.loadSBMLModel('https://www.ebi.ac.uk/biomodels/model/download/BIOMD0000000620.2?filename=BIOMD0000000620_url.xml')
```

```python
model.resetToOrigin()
print(model.getAntimony())
```

```python
results = model.simulate(0, 100, 1000)
model.plot(results)
```
